```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

# Задание 1

```python
# 1. С помощью библиотеки OpenCV считайте изображение:

image = cv2.imread('/content/saint-petersbourg-1936035_1280.jpg')

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.axis('off')  # Отключаем оси координат
plt.show()
```

```python
# 2. Постройте гистограмму черно-белого изображения:

gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap='gray')
```

"Построим" гистограмму (на самом деле — график гистограммы) черно-белого изображения. Гистограмма черно-белого изображения — это функция $h(n)$, принимающая целое число $n$ (интенсивность) в диапазоне $0 - 255$, и выдающая количество пикселей изображения с интенсивностью $n$.

```python
gray_hist = plt.hist(gray_img.ravel(), bins = 256)
plt.xlabel('Интенсивность $n$')
plt.ylabel('Количество $h(n)$')
plt.show()
```

```python
print("Максимальное значение функции:", np.max(gray_hist[0]))
print("Интенсивность при максимальном значении:", gray_hist[1][np.argmax(gray_hist[0])])
```

# Задание 2

```python
# 1. С помощью библиотеки OpenCV считайте изображение:

image = cv2.imread('/content/russia-2396022_1280.jpg')

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.axis('off')  # Отключаем оси координат
plt.show()
```

```python
# 2. Постройте гистограмму изображения в палитре RGB. Сравните оценку плотностей распределения интенсивностей по каналам R, G и B.

plt.hist(image_rgb[:, :, 0].ravel(), bins = 256, color = 'Red', alpha = 0.5, density=True)
plt.hist(image_rgb[:, :, 1].ravel(), bins = 256, color = 'Green', alpha = 0.5, density=True)
plt.hist(image_rgb[:, :, 2].ravel(), bins = 256, color = 'Blue', alpha = 0.5, density=True)
plt.hist(image_rgb.ravel(), bins = 256, color = 'orange', alpha = 0.5, density=True)
plt.legend(['Красный канал (R)', 'Зеленый канал (G)', 'Cиний канал (B)', 'Сумма по каналам'])
plt.xlabel('Интенсивность $n$')
plt.ylabel('Частота')
plt.show()
```

```python
# Находим максимальные значения плотности для каждого канала
max_density_red = np.max(np.histogram(image_rgb[:, :, 0].ravel(), bins=256, density=True)[0])
max_density_green = np.max(np.histogram(image_rgb[:, :, 1].ravel(), bins=256, density=True)[0])
max_density_blue = np.max(np.histogram(image_rgb[:, :, 2].ravel(), bins=256, density=True)[0])

max_density = max(max_density_red, max_density_green, max_density_blue)
print("Наибольшее значение плотности среди трех каналов:", max_density)
```

# Задание 3

```python
# 1. С помощью библиотеки OpenCV считайте изображение:

image = cv2.imread('/content/saint-petersbourg-1936035_1280.jpg')

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.axis('off')  # Отключаем оси координат
plt.show()
```

```python
# 2. Выполните линейную нормировку и растяжение на диапазон 0 - 255 гистограммы по каналам R, G и B.

image_rgb_new = image_rgb.copy()

image_rgb_new[:, :, 0] = (image_rgb_new[:, :, 0] - np.min(image_rgb_new[:, :, 0]))/(np.max(image_rgb_new[:, :, 0]) - np.min(image_rgb_new[:, :, 0]))*255
image_rgb_new[:, :, 1] = (image_rgb_new[:, :, 1] - np.min(image_rgb_new[:, :, 1]))/(np.max(image_rgb_new[:, :, 1]) - np.min(image_rgb_new[:, :, 1]))*255
image_rgb_new[:, :, 2] = (image_rgb_new[:, :, 2] - np.min(image_rgb_new[:, :, 2]))/(np.max(image_rgb_new[:, :, 2]) - np.min(image_rgb_new[:, :, 2]))*255

plt.imshow(image_rgb_new)
plt.axis('off')  # Отключаем оси координат
plt.show()
```

```python
# 3. Определите интенсивность каналов R, G, B для пикселя, находящегося на пересечении 437 строки и 671 столбца матричного представления изображения.

print('R,G,B:', image_rgb_new[437][671])
```
